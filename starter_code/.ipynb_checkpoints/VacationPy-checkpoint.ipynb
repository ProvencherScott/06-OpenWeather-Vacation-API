```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
#Create path to open the file
vacation_maps_data = "../starter_code/cities.csv"
```

```python
#Read the csv file as pandas
city_weather = pd.read_csv(vacation_maps_data)
#Create data frame with the columns shown in the table of the solution.
vacation_data = city_weather[["City", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Country", "Date"]]
vacation_data.head(30)
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#Store Lat and Lng as locations
locations = vacation_data[["Lat", "Lng"]]
#Store humidity as a series
humidity = vacation_data["Humidity"].astype(float)
```

```python
#Heatmap
fig = gmaps.figure(map_type="TERRAIN")
fig = gmaps.figure(center=(0, 0), zoom_level=2.25)

#Create a heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, max_intensity=100, point_radius=1)
fig.add_layer(heat_layer)

fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
vacation_data_cleaned = vacation_data.loc[((vacation_data["Max Temp"]<=85.0) & (vacation_data["Max Temp"]>73.0) & (vacation_data["Wind Speed"]<10.00) & (vacation_data["Cloudiness"]==0))]
vacation_data_cleaned.head(30)
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
#Store data frame above into variable
hotel_df = vacation_data_cleaned[["City", "Lat", "Lng", "Humidity", "Wind Speed", "Country"]]
#add column "Hotel Name"
hotel_df["Hotel Name"] = ""
hotel_df
```

```python
#Set parameters to search for hotels with 5000 meters using Google Places API

# the URL Sondra said we would be using for nearby search
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

params = {
    "radius": 5000,
    "types": "lodging",
    "keyword": "hotel",
    "key": g_key
}

# use iterrows to iterate through pandas dataframe
for index, row in hotel_df.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]
    
    params["location"] = f"{lat},{lng}"
    
    hotel_names = requests.get(base_url, params=params).json()
    
    try: 
        #Store the Hotel results into the "Hotel Name" column
        hotel_df.loc[index, "Hotel Name"] = hotel_names["results"][0]["name"]
    except (KeyError, IndexError):
        print(f"City {index} was not found")


    
    #cities_pd.loc[index, "Lat"] = cities_lat_lng["results"][0]["geometry"]["location"]["lat"]


```

```python
#Store the Hotel results into the "Hotel Name" column
hotel_df.head(20)
```

```python
hotel_df.dropna(inplace=True)
hotel_df.count()
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
marker_locations = hotel_df[["Lat", "Lng"]]
#adding the vacation_data to the variable
locations = vacation_data[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
#Heatmap
fig = gmaps.figure(map_type="TERRAIN")


heat_layer = gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, max_intensity=100, point_radius=1)
marker_layer = gmaps.marker_layer(marker_locations, info_box_content=hotel_info)
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)


fig
```
