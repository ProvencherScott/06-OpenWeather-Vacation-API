### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
#Create path to open the file
vacation_maps_data = "../starter_code/cities.csv"
```

```python
#Read the csv file as pandas
city_weather = pd.read_csv(vacation_maps_data)
#Create data frame with the columns shown in the table of the solution.
vacation_data = city_weather[["City", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Country", "Date"]]
vacation_data.head(30)
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#Store Lat and Lng as locations
locations = vacation_data[["Lat", "Lng"]]
#Store humidity as a series
humidity = vacation_data["Humidity"].astype(float)
```

```python
#Heatmap
fig = gmaps.figure(map_type="TERRAIN")
fig = gmaps.figure(center=(0, 0), zoom_level=2.25)

#Create a heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, max_intensity=100, point_radius=1)

fig.add_layer(heat_layer)

fig
```
